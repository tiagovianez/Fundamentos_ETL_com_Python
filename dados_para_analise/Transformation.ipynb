```python
import pandas as pd
import pandera as pa
```

```python
#Limpeza dos Dados
valores_ausentes = ['**', '***', '****', '*****', '###!', '###!', 'NULL']
#Leitura do arquivo e extração
df = pd.read_csv("ocorrencia.csv", parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)
df.head(10)
```

```python
#Criando a variável schema e atribuindo a ela um DataFrame do Pandera
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime, nullable=True),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int, nullable=True)  

    }
)
```

```python
#Teste de Validação
schema.validate(df)
```

```python
#Ocorrências cuja (classificação == INCIDENTE GRAVE ou classificação == INCIDENTE) e o estado == SP
#O comando 'isin' facilita a semântica e entendimento, por baixo ele está usando o operador 'or'
filtro1 = df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 & filtro2]
```

```python
#Montando novos modelos de dados e conteúdos que poderam ser utilizados em novas bases de dados.
#Filtrando ocorrencia_dia_hora entre: "11h do dia 03/12/2015 até 14h30 do dia 08/12/2015"
filtro1 = df.ocorrencia_dia_hora >= '2015-12-03 11:00:00'
filtro2 = df.ocorrencia_dia_hora <= '2015-12-08 14:30:00'
df.loc[filtro1 & filtro2]
```

```python
#--- AGRUPAMENTO ---
#ocorrências do ano de 2015 e mês 03
filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 3
#Determinando um novo DataFrame
df201503 = df.loc[filtro1 & filtro2]
df201503
```

```python
#Contando os dados do novo DataFrame
df201503.count()
```

```python
#Fazendo contagem de classificação através do codigo_ocorrencia
df201503.groupby(['ocorrencia_classificacao']).codigo_ocorrencia.count()
```

```python
#Fazendo contagem de classificação através do ocorrencia_aerodromo.
#Ps.: A coluna ocorrencia_aerodromo possui valores nulos, então não deverá ser efetuada a contagem, pois a informação virá distorcida.
df201503.groupby(['ocorrencia_classificacao']).ocorrencia_aerodromo.count()
```

```python
#Ao invés de contar por uma coluna, conte por registros 'size'
#Não será preciso informar uma coluna, será contado apenas o número de linhas que foram agrupadas.
#Sem ter o risco de possuir valores nulos na hora da contagem
df201503.groupby(['ocorrencia_classificacao']).size()
```

```python
#Imprimindo em ordem crescente
#Transformando dados brutos em informações
df201503.groupby(['ocorrencia_classificacao']).size().sort_values()
```

```python
#Imprimindo em ordem descrescente
#Transformando dados brutos em informações
df201503.groupby(['ocorrencia_classificacao']).size().sort_values(ascending=False)
```

```python
#Obtendo agrupamento dos dados através da região 'SUDESTE' no ano de '2010'
filtro1 = df.ocorrencia_dia.dt.year == 2010
filtro2 = df.ocorrencia_uf.isin(['SP', 'RJ', 'MG', 'ES'])
dfsudeste2010 = df.loc[filtro1 & filtro2]
dfsudeste2010
```

```python
#Sobre o novo DataFrame, irei gerar novos agrupamentos de dados
#Gerando o número de ocorrências dos acidentes ocorridos na região sudeste em 2010 por estado
dfsudeste2010.groupby(['ocorrencia_uf', 'ocorrencia_classificacao']).size()
```

```python
#Gerando o número de ocorrências nas cidades da região sudeste em 2010
dfsudeste2010.groupby(['ocorrencia_cidade']).size().sort_values(ascending=False)
```

```python
filtro = dfsudeste2010.ocorrencia_cidade == 'RIO DE JANEIRO'
dfsudeste2010.loc[filtro]
```

```python
#Total de recomendacoes da cidade do rio de janeiro
filtro = dfsudeste2010.ocorrencia_cidade == 'RIO DE JANEIRO'
dfsudeste2010.loc[filtro].total_recomendacoes.sum()
```

```python
#Relacionando o número de recomendações por cidade
dfsudeste2010.groupby(['ocorrencia_cidade']).total_recomendacoes.sum()
```

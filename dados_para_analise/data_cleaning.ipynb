```python
#A função replace permite que faz alterações no dataset como todo
#Ps.: a função 'inplace=True', permite que a alteração ocorra de fato
df.replace(['**', '****', '*****', '###!', '###!', 'NULL'], pd.NA, inplace=True)
```

```python
#Contabilizando os dados que não foram informados (NA)
df.isna().sum()
```

```python
#Colocando valor nos dados não informados (Em tempo de visualização) 
df.fillna(10, inplace=True)
```

```python
df.head()
```

```python
#Revertendo o valor que estava em 10 para NA (NotAvaiable)
#É importante saber quais dados não informados estamos trabalhando
df.replace([10], pd.NA, inplace=True)
```

```python
df.isna().sum()
```

```python
#Alterando o valor de apenas uma coluna.
#Ps.: Lembres-se que a alteração só ocorrerá de fato quando se utilizar o comando 'inplace=True'
df.fillna(value={'total_recomendacoes':10}, inplace=True)
```

```python
df.isna().sum()
```

```python
#Criando um backupde uma coluna
df['total_recomendacoes_bkp'] = df.total_recomendacoes
df
```

```python
#Removendo uma coluna "df.drop"| "axis=1" representa o eixo da coluna
df.drop(['total_recomendacoes_bkp'], axis=1, inplace=True)
df.head()
```

```python
#Excluindo todos os valores não informados. Ps.: Ele remove as linhas não as células.
df.dropna()
```

```python
#Dropando o NA por coluna, ao invés de excluir várias linhas.
df.dropna(subset=['ocorrencia_uf'])
df
```
